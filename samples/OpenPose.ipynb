```python
import sys
import cv2
from matplotlib import pyplot as plt
```

```python
openpose_path = "/workspace/openpose/build/python"
if openpose_path not in sys.path:
    sys.path.append(openpose_path)
from openpose import pyopenpose as op
```

```python
# Starting OpenPose
params = {"model_folder": "/workspace/openpose/models"}
opWrapper = op.WrapperPython()
opWrapper.configure(params)
opWrapper.start()
```

```python
path = "/workspace/body-shape-detector/api/images/img_01.jpg"
datum = op.Datum()
input_image = cv2.imread(path)
datum.cvInputData = input_image
opWrapper.emplaceAndPop(op.VectorDatum([datum]))
network_output = datum.poseKeypoints
```
